```python
import cv2 
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
test_img = np.zeros((640,480,3),dtype=np.uint8)
test_img = cv2.circle(test_img,(150,150), 20, (0,255,0),-1)

from_corners = np.array([[100,100],[300,100],[300,600],[100,600]],dtype=np.float32)
to_corners = np.array([[0,0],[200,0],[200,500],[0,500]],dtype=np.float32)
dsize = (200,500)

transform_matrix = cv2.getPerspectiveTransform(from_corners,to_corners)

x = 150
y = 150

img_bin = test_img[:,:,1]
plt.imshow(test_img)
plt.show()
```

```python
np.dot(transform_matrix[0:,],np.array((x,y,1)))
```

```python
out = cv2.warpPerspective(test_img, transform_matrix, dsize)
plt.imshow(out)
plt.show()
```

```python
timeit cv2.moments(img_bin)
```

```python
timeit np.sum(img_bin)
```

```python
hsv_min_1 = (0, 20, 173)
hsv_max_1 = (11, 88, 195)
hsv_min_2 = (172, 20, 173)
hsv_max_2 = (181, 88, 195)

rgb_vals = []
for max_, min_ in zip((hsv_max_1,hsv_max_2),(hsv_min_1,hsv_min_2)):
    h_vals = range(min_[0],max_[0])
    s_vals = range(min_[1],max_[1])  
    v_vals = range(min_[2],max_[2])
    for h in h_vals:
        for s in s_vals:
            for v in v_vals:
                rgb_vals.append(cv2.cvtColor(np.array([[[h,s,v]]],dtype=np.uint8),cv2.COLOR_HSV2RGB))

rgb_vals = np.squeeze(np.array(rgb_vals))
plt.hist(rgb_vals[:,0],label='r',color = 'r')
plt.hist(rgb_vals[:,1],label='g',color = 'g')
plt.hist(rgb_vals[:,2],label='b',color = 'b')
plt.show()

for i in range(3):
    print((min(rgb_vals[:,i]),max(rgb_vals[:,i])))
lowerb = np.array(list(reversed(list((min(rgb_vals[:,i]) for i in range(3))))),dtype=np.uint8)
upperb = np.array(list(reversed(list(max(rgb_vals[:,i]) for i in range(3)))),dtype=np.uint8)
```

```python
print(upperb,lowerb)
```

```python
%matplotlib inline
testig = cv2.imread("testFrame.png")

plt.imshow(testig)
plt.show()

test_img_hsv = cv2.cvtColor(testig,cv2.COLOR_BGR2HSV)
hsv_min_1 = (0, 20, 173)
hsv_max_1 = (11, 88, 195)
hsv_min_2 = (172, 20, 173)
hsv_max_2 = (181, 88, 195)
low_hue_bin_img = (cv2.inRange(test_img_hsv, hsv_min_1, hsv_max_1)==255)
high_hue_bin_img = (cv2.inRange(test_img_hsv, hsv_min_2, hsv_max_2)==255)
bin_img_hsv = (low_hue_bin_img) | (high_hue_bin_img)

bin_img = cv2.inRange(testig, lowerb,upperb)
plt.imshow(bin_img)
plt.show()
plt.imshow(bin_img_hsv)
plt.show()
```
